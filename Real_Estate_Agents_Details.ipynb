```python
url = 'https://www.google.com/maps/search/'+'Real Estate agents in Adityapur'
bot = Bot()
bot.driver.get(url)
```

```python
Title = []
Address = []
Phone = []
```

```python
import time
while True:
    time.sleep(3)
    WebDriverWait(bot.driver,20).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME,'section-result-text-content')))
    for i in bot.driver.find_elements_by_class_name('section-result-text-content'):
        txt = i.text
        try:
            li = txt.split('\n')
            Title.append(li[0])
            if len(li)==1:
                Phone.append('No Number')
                Address.append('No Address')
                continue
            Phone.append(int(''.join(li[-1][-12:].split(' '))))
            Address.append(li[-2])
        except:
        
            if len(li[1])>10:
                Address.append(li[1])
            else:
                Address.append(li[-1])
            Phone.append('No Number')
    try:
        WebDriverWait(bot.driver,20).until(
            EC.element_to_be_clickable((By.ID,'n7lv7yjyC35__section-pagination-button-next'))).click()
    except:
        break
    
        
```
